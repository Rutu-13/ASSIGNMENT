```python
Measure_x =pd.Series(Measure_x)
```

```python
Measure_x.describe()
```

```python
Q1 = np.quantile(Measure_x,0.25)
Q1
```

```python
Q3 = np.quantile(Measure_x,0.75)
Q3
```

```python
M=Measure_x.median()
```

```python
IQR = Q3-Q1 
IQR
```

```python
upper_bond = Q3+(1.5*IQR)
upper_bond
```

```python
lower_bond = Q1+(1.5*IQR)
lower_bond
```

```python
print('First Quantile=', Q1, 'Second Quantile=', M, 'Third Quantile=', Q3,
      'Inter-Quartile Range=', IQR, 'Upper Whisker=', upper_bond, 'Lower Whisker=', lower_bond)
```

```python
Outliers = Measure_x[(Measure_x <= lower_bond) | (Measure_x >= upper_bond)]
print('The outlier in the boxplot:',Outliers)
```

```python
Outliers = Measure_x[(Measure_x >= upper_bond)]
print('The outlier in the boxplot:',Outliers)
```
