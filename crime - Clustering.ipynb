```python
crime = pd.read_csv(r"C:\Users\DELL\OneDrive\Desktop\assignment\7\crime_data.csv")
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_crime = scaler.fit_transform(crime.iloc[:,1:])
```

```python
#Normalized data fuction
def norm_func(i):
    x=(i-i.min())/(i.max()-i.min())
    return(x)
```

```python
df_norm1=norm_func(crime.iloc[:,1:])
```

```python
# elbow curve
wcss=[]
for i in range(1,11):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(df_norm1)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('Elbow curv')
plt.xlabel('number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Build cluster plot
clus_new = KMeans(4, random_state=42)
clus_new.fit(scaled_crime)
```

```python
# assign cluster 
crime['clus_id']=clus_new.labels_
clus_new.cluster_centers_
```

```python
crime.groupby('clus_id').agg(['mean']).reset_index()
```

```python
crime
```

```python
kmeans = KMeans(n_clusters=4, init= 'k-means++', random_state=42)
y_kmeans = kmeans.fit_predict(scaled_crime)
```

```python
plt.scatter(scaled_crime[y_kmeans == 0, 0], scaled_crime[y_kmeans == 0, 1], s = 50, c = 'magenta', label = 'Cluster 1')
plt.scatter(scaled_crime[y_kmeans == 1, 0], scaled_crime[y_kmeans == 1, 1], s = 50, c = 'blue', label = 'Cluster 2')
plt.scatter(scaled_crime[y_kmeans == 2, 0], scaled_crime[y_kmeans == 2, 1], s = 50, c = 'green', label = 'Cluster 3')
plt.scatter(scaled_crime[y_kmeans == 3, 0], scaled_crime[y_kmeans == 3, 1], s = 50, c = 'cyan', label = 'Cluster 4')

plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 100, c = 'yellow', label = 'centriod')

plt.title('Clusters of crime')
plt.xlabel('Annual crime record')
plt.ylabel('Score')
plt.legend()
plt.show()
```

```python
df_c=df.drop(columns=['Place'])
df_c
```

```python
df_array = df_c.values
df_array
```

```python
#normalize data
from sklearn.preprocessing import StandardScaler
st=StandardScaler().fit(df_c)
norm_d=st.transform(df_c)
```

```python
#DBSCAN clustering
from sklearn.cluster import DBSCAN
df_model = DBSCAN(eps=1.2,min_samples=6).fit(norm_d)
df_model
```

```python
#Noisy samples are given the label -1.
df_model.labels_
```

```python
df_clust = pd.DataFrame(df_model.labels_,columns=['clusters'])
df_clust
```

```python
df_clust.value_counts()
```

```python
#-1 cluster represents outliers which means we have 18 data points as outliers
final_data=pd.concat([df_clust,df_c],axis=1)
final_data
```
