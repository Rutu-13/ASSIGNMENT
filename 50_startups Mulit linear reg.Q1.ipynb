```python
# calculate VTF
rs_r=smf.ols('RDS~ADMS+MKS',data=st1).fit().rsquared
vif_r = 1/(1-rs_r)

rs_a=smf.ols('ADMS~RDS+MKS',data=st1).fit().rsquared
vif_a = 1/(1-rs_a)

rs_m=smf.ols('MKS~ADMS+RDS',data=st1).fit().rsquared
vif_m = 1/(1-rs_m)

#putting values in dataframe format
df={'Variables':['RDS','ADMS','MKS'],'vif':[vif_r,vif_a,vif_m]}
vif_d=pd.DataFrame(df)
vif_d
```

```python
while np.max(c)>0.5:
    st_m = smf.ols("Profit~RDS+ADMS+MKS",data=st2).fit()
    (c,_)=st_m.get_influence().cooks_distance
    c
    np.argmax(c) , np.max(c)
    st2=st2.drop(st2.index[[np.argmax(c)]],axis=0).reset_index(drop=True)
    st2
else:
    final_m=smf.ols("Profit~RDS+ADMS+MKS",data=st2).fit()
    final_m.rsquared , final_m.aic
    print("Thus model accuracy is improved to",final_m.rsquared)
```
